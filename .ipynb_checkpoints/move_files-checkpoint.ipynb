```python
import os
import random
import shutil
```

#### Move Lying Down

```python
# Source and destination folders
source_folder = 'C:\\Users\\johnl\\Documents\\GitHub\\WTH_StepBros\\train_lyingdown'
destination_folder = 'C:\\Users\\johnl\\Documents\\GitHub\\WTH_StepBros\\test_lyingdown'

# Number of random images to select and move
#Total number of lying down images: 235. 0.8 = 188. 0.2 = 47
num_images_to_select = 47  

# Get a list of all image files in the source folder
image_files = [f for f in os.listdir(source_folder) if f.lower().endswith(('.jpg',))]

# Select random images
selected_images = random.sample(image_files, num_images_to_select)

# Move selected images to the destination folder
for image in selected_images:
    source_path = os.path.join(source_folder, image)
    destination_path = os.path.join(destination_folder, image)
    shutil.move(source_path, destination_path)
    print(f"Moved {image} to {destination_folder}")
```
